```python
# Membuat Data Frame
# diwali dengan pd.DataFrame
# didalamnya bisa menggunakan dictionary atau multidimensi data list
pd.DataFrame({
    'nama' : ['ayah', 'ibu', 'kakak', 'saya', 'adik'],
    'umur' : [50, 40, 20, 15, 10],
    'makanan_favorite' : ['ayam', 'bakso', 'mie', 'kfc', '']
})
```

```python
pd.DataFrame({
    'nama' : ['ayah', 'ibu', 'kakak', 'saya', 'adik'],
    'umur' : [50, 40, 20, 15, None],
    'makanan_favorite' : ['ayam', 'bakso', 'mie', 'kfc', np.nan]
})
```

string kosong '' berbeda dengan None atau np.nan
- None / np.nan = tidak ada data (missing)
- ' ' = datanya ada tapi hanya string kosong

Aturan membuat Data Frame:
- value dari dictionary harus sebuah list
- jumlah elemen harus sama

```python
# Challenge
# tolong carikan pelanggan yang lahirnya bukan di bulan Mei
df_customers[df_customers['bulan_lahir'] != 'Mei']
```

```python
# tolong carikan pelanggan yang lahirnya di bulan April, Mei, Juni
# menggunakan fungsi yang sama dengan SQL
# bis amenggunakan multiple condition
df_customers[df_customers['bulan_lahir'].isin(['April','Mei', 'Juni'])]
```

### Kesamaan dengan SQL Condition

```python
# WHERE telepon IS NOT NULL
df_customers[df_customers['telepon'].notnull()].head()
```

```python
# WHERE 'column' IS NULL
df_customers[df_customers['bulan_lahir'].isnull()].head()
```

```python
# WHERE tanggal_registrasi LIKE '2012'
# ^depan string
# $belakang string
df_customers[df_customers['tanggal_registrasi'].str.contains('2012')].head()
```

## Multiple Condition

```python
df_customers[(df_customers['umur'] < 30) & (df_customers['bulan_lahir'] == 'Juni')]
```

```python
df_customers[(df_customers['umur'] < 30) & (df_customers['bulan_lahir'].isin(['Juni', 'Juli']))]
```

```python
# challange berhadiah
# tolong carikan customers yang lahir bukan di bulan Januari
# dia menggunakan gmail
# namanya diawal dengan huruf R

df_customers[
    (df_customers['bulan_lahir'] != 'Januari') &
    (df_customers['email'].str.endswith('@gmail.com')) &
    (df_customers['nama'].str.startswith('R'))
]
```

```python
# cara lain
df_customers[
    (df_customers['bulan_lahir'] != 'Januari') &
    (df_customers['email'].str.contains('gmail')) &
    (df_customers['nama'].str.contains('^R'))
]
```
